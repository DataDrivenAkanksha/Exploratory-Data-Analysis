Q17. Create a new column to show the percentage of total runs scored in 4s and 6s. Then print the top 5 players with maximum percentage.

```python
# Calculate the percentage of total runs from boundaries (4s & 6s)
ipl_df["pecentage_of_total_runs"] = ((ipl_df['4s']*4 + ipl_df["6s"]*6)/ipl_df["Total_Runs"])*100

# Group by Player_Name and find the top 5 players with the highest percentage of boundary runs
Top_5_boundary_scorers = ipl_df.groupby("Player_Name")["pecentage_of_total_runs"].max().sort_values(ascending=False).head(5)

# Display results
print(Top_5_boundary_scorers)
```

Q18. Print the players with top 5 Not out percentages (Not Out percentage can be calculated as number of Not outs divided by Innings).

```python
# Calculate the Not out percentages
ipl_df["not_out_percentage"] = ipl_df.apply(lambda x: (x['Not_Outs'] / x['Innings_Batted']) * 100 if x['Innings_Batted'] > 0 else 0, axis=1)

# Group by Player_Name and find the top 5 players with the highest not out percentage
Top_5_not_out_percentage = ipl_df.groupby("Player_Name")["not_out_percentage"].max().sort_values(ascending=False).head(5)

# Display results
print(Top_5_not_out_percentage)
```
